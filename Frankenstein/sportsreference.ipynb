# Previous Games Stats

These two functions work together to return a new DataFrame() with the previous 20 games stats.

```python
# get pertinent stats from the previous 20 games
## game_i corresponds to the i-th most recent game

def get_previous_game(team_data, current_game_number):
    previous_game_number = current_game_number - 1 # start with values over 20
    return team_data.loc[previous_game_number, ]

def get_previous_20_games(team_data, current_game_number):
    previous_20 = pd.DataFrame()
    n = 20
    while n > 0:
        previous_20 = pd.concat([previous_20, 
                                 get_previous_game(team_data, current_game_number-n)],
                               axis=1)
        n -= 1
    
    return previous_20.T.reset_index()
    
```

```python
'''
For my own future reference, I thought a verbose description of
what I will call, Unique Game IDs (UGID), for the remainder of
this project would be important:

UGIDs have the following format:
    AAAYYYYMMDDN
        EX: ANA201506090
    
These can all be found from the game log files.

This way all games have a unique identifier and ensures that there are no duplicates

further explanation of each field:
    AAA - This is the 3 letter RetroSheet abbreviation for a team.
        NOTE: These are not the same as the common team abbreviations.
        EX: Los Angeles Angels (ANA)
            FUN FACT: the 3 letter abbreviation for the Angels is ANA in this
            case b/c it actually stands for Anaheim. If you are interested
            in why the franchises are named the way they are, look here:
            https://www.retrosheet.org/boxesetc/MISC/FRDIR.htm
    YYYYMMDD - This is the fields representing the date which the game occured.
        YYYY represents the 4-digit year, MM represent the 2-digit month,
        and DD represents the 2-digit day.
        EX: 20150609
    N - This field repressents the specific game from the day the game was played.
        NOTE: This field can only take 3 values: 0, 1 and 2, explained below:
            0 - only a single game was played on day YYYYMMDD
            1 - first game of a double-header on day YYYYMMDD
            2 - second game of a double header on day YYYYMMDD
        EX: 0
        
Furthermore, an example is given below to show how to query a game log and concatenate a UGID.
The batting statistics for a team will be found in the HOME TEAMS .EV* file in datasets/retro_sheet_pbp directory

IMPORTANT: the AAA field must always be the abbreviation of the HOME TEAM. Trying to use a 
    visiting team's abbreviation will result in an ERROR!
'''


previous_game_id = (pit_previous_20.loc[19,'Home Team'] # use HOME TEAM abbreviation!
                    + str(pit_previous_20.loc[19,'Date']) 
                    + str(pit_previous_20.loc[19,'Number of game']))

previous_game_id
```

## Putting it All Together...

Now I define a function to take in a data set (RetroSheet game logs) and calculate the BA for the starters of each game. A more complete description of the function itself follows in the multiline comment, so i'll save some time and not type it again here. Of course, LMK if something from here or above is not clear and I'll revise the comments.

```python
'''
Whew! This value is exactly the value that we got 
from using the sportsreferences BoxScore() object!

Now we can attempt to generalize and automate this process by
defining a function to perform this for us.

This function will take in a RetroSheet DataFrame(), along with several options
and update the hits and at-bat fields for the both the home and visiting
team. Here is a description of it's prototype, parameters and what it'll return:

batting_avg(data, n=20, file_path='.', output='silent')

parameters:
    data - The dataframe containing the RetroSheet data
    n [default = 20] - Number of previous games to compute ba for. If you're interested
        in computing the ba far all games in the data set, one possible way to do
        this would be to pass DataFrame().shape[0]
    file_path [default = '.'] - The relative or absolute file path to the .EV* data
    output - 3 options:
        'silent' [default] - Function will return a DataFrame() (an updated version of the supplied
            data).
        'test' - Function will return two List(), one for the visiting team's
            ba nad on for the home team's ba. No dataframe will be returned.
        'verbose' - Function will perform as if you used 'silent' and will
            print additional info that is helpful for debugging

return:
    This function may return a DataFrame() or List(). See output parameter
    for more details.
    
I'll be experimenting with more ways to speed up this process, but
for now this shall suffice. This may also work well for calculating 
other stats, too... may refactor this in the future to accomodate.
'''

# I define file_path to be the current working directory (cwd) by default
# I recommend that a file path is ALWAYS supplied, since I'm
# assuming most people will not be using this file and
# data files in the same directory. I wanted to provide a 
# default value just in case.
# 
# maybe if I move all of these function into a module (hmmmmmm)
# then this can be streamlined....
def batting_avg(data, file_path='.', output='silent'):
    # these list are used solely for testing and will only be returned
    # if output='test'
    visiting_ba = []
    home_ba = []
    # Additionally, if output='verbose', this function will
    # perform in 'test' mode as well as print info
    
    for i in range(data.shape[0]):
        # step 1: create variables
        date = str(data.iloc[i].loc['Date']) # format - YYYYMMDD
        year = date[0:4]
        home_team = data.loc[i,'Home Team']
        home_league = data.loc[i, 'Home Team League']
        game_number = str(data.loc[i, 'Number of game'])
        if home_league == 'NL':
            home_league = 'N'
        else:
            home_league = 'A'
        file = year + home_team + '.EV' + home_league
        
        # step 2: read in home teams batting data
        batting_data = pd.read_csv(os.path.join(file_path, file))
        
        # step 3: find UGID - AAAYYYYMMDDN
        ugid = home_team + date + game_number
        
        # step 4: filter current game batting data
        current_game_batting_data = batting_data[batting_data['game_id'] == ugid]
        
        # step 5: calculate visiting team ba
        visitor_hits, visitor_at_bats = 0, 0

        for batter in [data.loc[i,'Visiting Team Player ' + str(j+1) + ' ID'] for j in range(9)]:
            batter_stats = current_game_batting_data[current_game_batting_data['player_id'] == batter]
            visitor_hits += batter_stats[batter_stats['hit_value'] > 0].shape[0]
            visitor_at_bats += batter_stats[batter_stats['ab_flag'] == 'T'].shape[0]
            
        # step 6: calculate home team ba
        home_hits, home_at_bats = 0, 0

        for batter in [data.loc[i,'Home Team Player ' + str(j+1) + ' ID'] for j in range(9)]:
            batter_stats = current_game_batting_data[current_game_batting_data['player_id'] == batter]
            home_hits += batter_stats[batter_stats['hit_value'] > 0].shape[0]
            home_at_bats += batter_stats[batter_stats['ab_flag'] == 'T'].shape[0]
            
        # step 7: update old data
        data.loc[i, 'Visiting Team Hits'] = visitor_hits
        data.loc[i, 'Visiting Team At-bats'] = visitor_at_bats
        
        data.loc[i, 'Home Team Hits'] = home_hits
        data.loc[i, 'Home Team At-bats'] = home_at_bats
        
        # step 8 (conditional): testing
        # add additional print statements for whatever you're interested in
        if output == 'test' or output == 'verbose':
            if output == 'verbose':
                print(visitor_hits)
                print(visitor_at_bats)
                print(ugid)
                print(file)
            visiting_ba.append(visitor_hits/visitor_at_bats)
            home_ba.append(home_hits/home_at_bats)
            
    if output == 'test' or output == 'verbose':   
        return visiting_ba, home_ba
    
    return data
```

```python
def batting_avg_2(gl, batting, output='silent'):
    # these list are used solely for testing and will only be returned
    # if output='test'
    visiting_ba = []
    home_ba = []
    # Additionally, if output='verbose', this function will
    # perform in 'test' mode as well as print info
    
    for i in range(gl.shape[0]):
        # step 1: create variables
        date = str(gl.iloc[i].loc['Date']) # format - YYYYMMDD
        year = gl[0:4]
        home_team = gl.loc[i,'Home Team']
        home_league = gl.loc[i, 'Home Team League']
        game_number = str(gl.loc[i, 'Number of game'])
        if home_league == 'NL':
            home_league = 'N'
        else:
            home_league = 'A'
        
        # step 3: find UGID - AAAYYYYMMDDN
        ugid = home_team + date + game_number
        
        # step 4: filter current game batting data
        current_game_batting_data = batting[batting['game_id'] == ugid]
        
        # step 5: calculate visiting team ba
        visitor_hits, visitor_at_bats = 0, 0

        for batter in [gl.loc[i,'Visiting Team Player ' + str(j+1) + ' ID'] for j in range(9)]:
            batter_stats = current_game_batting_data[current_game_batting_data['player_id'] == batter]
            visitor_hits += batter_stats[batter_stats['hit_value'] > 0].shape[0]
            visitor_at_bats += batter_stats[batter_stats['ab_flag'] == 'T'].shape[0]
            
        # step 6: calculate home team ba
        home_hits, home_at_bats = 0, 0

        for batter in [gl.loc[i,'Home Team Player ' + str(j+1) + ' ID'] for j in range(9)]:
            batter_stats = current_game_batting_data[current_game_batting_data['player_id'] == batter]
            home_hits += batter_stats[batter_stats['hit_value'] > 0].shape[0]
            home_at_bats += batter_stats[batter_stats['ab_flag'] == 'T'].shape[0]
            
        # step 7: update old data
        gl.loc[i, 'Visiting Team Hits'] = visitor_hits
        gl.loc[i, 'Visiting Team At-bats'] = visitor_at_bats
        
        gl.loc[i, 'Home Team Hits'] = home_hits
        gl.loc[i, 'Home Team At-bats'] = home_at_bats
        
        # step 8 (conditional): testing
        # add additional print statements for whatever you're interested in
        if output == 'test' or output == 'verbose':
            if output == 'verbose':
                print(i)
            visiting_ba.append(visitor_hits/visitor_at_bats)
            home_ba.append(home_hits/home_at_bats)
            
    if output == 'test' or output == 'verbose':   
        return visiting_ba, home_ba
    
    return gl
```
