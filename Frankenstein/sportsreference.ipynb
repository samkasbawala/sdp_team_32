```python
# get the team team_abrv boxscores from data

def get_team(data, team_abrv):
    return data[(data['Visiting Team'] == team_abrv) | (data['Home Team'] == team_abrv)]
```

```python
pit_2015 = get_team(baseball_games_2015, team_abrv='PIT')
```

```python
pit_2015.reset_index(inplace=True)
pit_2015.head()
```

```python
# get pertinent stats from the previous 20 games
## game_i corresponds to the i-th most recent game

def get_previous_game(team_data, current_game_number):
    previous_game_number = current_game_number - 1 # start with values over 20
    return team_data.loc[previous_game_number, ]

def get_previous_20_games(team_data, current_game_number):
    previous_20 = pd.DataFrame()
    n = 20
    while n > 0:
        previous_20 = pd.concat([previous_20, 
                                 get_previous_game(team_data, current_game_number-n)],
                               axis=1)
        n -= 1
    
    return previous_20.T.reset_index()
    
```
