```python
from sportsreference.mlb.schedule import Schedule

houston_schedule = Schedule('HOU')
i = 5
for game in houston_schedule:
    print(game.date)
    print(game.result)
    for player in game.boxscore.away_players:
        print(player.name)
    if i == 0:
        break
    i -= 1
```

```python
hou_schedule = Schedule('HOU')

game = hou_schedule[0]
date = game.datetime
print(date)
print(game.day_or_night)

player = game.boxscore.home_players[0]
print(player.name, "(BIG UConn guy)")
print("hits:", player.hits)
print("at bats:", player.at_bats)
print("plate appearances:", player.plate_appearances)
print("strikeouts:", player.times_struck_out)
print("batting average:", player.batting_average)
print("RBI:", player.runs_batted_in)
print("runs scored:", player.runs)
```
